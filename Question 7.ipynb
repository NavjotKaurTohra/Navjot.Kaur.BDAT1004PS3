# Question 7
Create a vertical bar chart comparing the number of marriages and divorces per 
capita in the U.S. between 1900, 1950, and 2000.
Don't forget to label your axes!

```python
import pandas as pd

marriage_divorces = pd.read_csv('us-marriages-divorces-1867-2014.csv')


marriage_divorces_filter = marriage_divorces[marriage_divorces['Year'].apply(lambda x: x in [1900, 1950, 2000])]

years = marriage_divorces_filter['Year'].values
marriages_per_capita = marriage_divorces_filter['Marriages_per_1000'].values
divorces_per_capita = marriage_divorces_filter['Divorces_per_1000'].values


df = pd.DataFrame(
   dict(
      marriages_per_capita=marriages_per_capita,
      years=years
   )
)

import matplotlib.pyplot as plt


plt.bar(marriages_per_capita,divorces_per_capita)


for x,y,a in zip(marriages_per_capita,divorces_per_capita,years):

    label = "{:.2f}".format(y)

    plt.annotate(label, # this is the text
                 (x,y), # these are the coordinates to position the label
                 textcoords="offset points", # how to position the text
                 xytext=(0,10), # distance from text to points (x,y)
                 ha='center')# horizontal alignment can be left, right or center    
    plt.annotate(a,(x,y-0.5))



plt.title('Number of marriages and divorces per capita in the U.S.')
plt.ylabel('Divorces_per_1000')
plt.xlabel('marriages_per_capita')
plt.show()

```
