# Question 10
Create a scatter plot showing the relationship between the total revenue earned by 
arcades and the number of Computer Science PhDs awarded in the U.S. between 
2000 and 2009.
Don't forget to label your axes!
Color each dot according to its year

```python
import matplotlib.pyplot as plt

import pandas as pd
arcade_revenue_cs_doctorates = pd.read_csv('arcade-revenue-vs-cs-doctorates.csv')
arcade_revenue = arcade_revenue_cs_doctorates['Total Arcade Revenue (billions)'].values
cs_doctorates_awarded = arcade_revenue_cs_doctorates['Computer Science Doctorates Awarded (US)'].values

colors=["#0000FF", "#00FF00", "#FF0066","#FF0067","#FF0766",   "#0000FA", "#00FFA0", "#FF0166","#FF00B7","#FB0766"]

for i in range(len(arcade_revenue)):
    plt.scatter(arcade_revenue[i], cs_doctorates_awarded[i], color=colors[i])


plt.title("Total revenue earned by arcades V/s the number of Computer Science PhDs awarded in US")
plt.xlabel("Total revenue earned by arcades")
plt.ylabel("Number of Computer Science PhDs awarded")

plt.show() 
```
