# Question 6
Create a line plot showing the number of marriages and divorces per capita in the 
U.S. between 1867 and 2014. Label both lines and show the legend.
Don't forget to label your axes!

```python
import pandas as pd

marriage_divorce = pd.read_csv('us-marriages-divorces-1867-2014.csv')

marriage_divorce_filter = marriage_divorce[marriage_divorce['Year'].apply(lambda x: x >=1867 and x <= 2014)]

years = marriage_divorce_filter['Year'].values
marriages = marriage_divorce_filter['Marriages'].values
divorces_per_capita = marriage_divorce_filter['Divorces_per_1000'].values

import matplotlib.pyplot as plt

plt.plot(divorces_per_capita, marriages, marker='o', label='years')
plt.title('Marriages Vs divorces_per_capita', fontsize=14)
plt.xlabel('marriages', fontsize=14)
plt.ylabel('divorces_per_capita', fontsize=14)
plt.legend()
plt.grid(True)
plt.show()


```
